### 2 РАССМОТРИМ, КАК МЕНЯЛИСЬ ПРОДАЖИ ПО ПЛАТФОРМАМ

Рассмотрим самые прибыльные платформы

```python
#Определим самые прибыльные платформы
pivot_data = check_data.pivot_table(index='platform', values='total_sales', aggfunc='sum')
pivot_data.sort_values('total_sales', ascending=False).plot(kind='bar', title='Продажи по платформам')
```

По гистаграмме видно, что самые прибыльные платформы это PS2, Xbox360, PS3, WII, DS и PS. Так как PS совсем старая платформа, возьмем первые 5 платформ и построим распределение продаж на этих платформах по годам за последние годы

```python
#Возьмем первые 5 платформ как самые прибыльные и постоим распредление продаж по годам
top5_pivot_data = pivot_data.sort_values('total_sales', ascending=False).query('total_sales  > 800')
top5_pivot_data.index

#pivot_game_data.plot(title='распределение Наиболее прибыльных игр по годам').legend()
```

```python
#Получаем данные игр для нужных платформ
top_game_data = check_data.query('platform in @top5_pivot_data.index and year_of_release > 2000')
```

```python
#Строим график 
plt.figure(figsize=(12,6))
plt.title('Количество продаж популярных игровых платформ')
plt.xlabel('Годы игровых релизов')
plt.ylabel('Продажи')
for platform in top5_pivot_data.index:
    platform_data = top_game_data[top_game_data['platform'] == platform].pivot_table(index='year_of_release', 
                                                                                     values='total_sales', 
                                                                                     aggfunc='sum')
    sns.lineplot(data=platform_data['total_sales'], label=platform)
    plt.legend()
```

По графику видно, что PS2 продажи игр для PS2 падают с 2004 года, так как появилась PS3. 
По графику продаж игр для PS3 и XBOX360 можно понять, что актуальность платформы составляет примерно 10 лет. Пик продаж наступает на середине(Примерно через 5 лет после запуска приставки). По графику видно, что на момент 2016 года все эти платформы уже не актуальны, но их данные полезны, чтобы понимать, через какое время после выпуска платформы можно рассчитывать на максимальные продажи

```python
#Приведем сводную таюлицу для определения продолжительности актуальности платформы
ppivot_top_game_data = top_game_data.pivot_table(index='platform',
                                                 values='year_of_release',
                                                 aggfunc=['min', 'max'])
ppivot_top_game_data['sales_years'] = ppivot_top_game_data['max'] - ppivot_top_game_data['min']
ppivot_top_game_data.columns = ['first_release_year', 'last_release_year', 'sales_years']
ppivot_top_game_data
```

### 7 Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами

```python
#Напишем функцию, которая вывод диаграммы рассеяния для каждой популярной платформы и считает корреляцию
def data_platform(platform):
    platform_data = top5_game_data[top5_game_data['platform'] == platform]
    fig, ax = plt.subplots(1, 2, figsize=(18, 6))
    sns.scatterplot(x='user_score', y='total_sales', data=platform_data, ax=ax[0])
    sns.scatterplot(x='critic_score', y='total_sales', data=platform_data, ax=ax[1])
    fig.suptitle(platform)
    ax[0].set(xlabel='Оценка пользователей')
    ax[1].set(xlabel='Оценка критиков')
    ax[0].set(ylabel='Количество продаж')
    ax[1].set(ylabel='Количество продаж')
    plt.show()
    
    user_correl = platform_data['user_score'].corr(platform_data['total_sales'])
    critic_correl = platform_data['critic_score'].corr(platform_data['total_sales'])
    
    if 0 < user_correl <= 0.3:
        print('Корреляция между оценкой пользователей и продажами слабая и равна', user_correl)
    if 0.3 < user_correl <= 0.7:
        print('Корреляция между оценкой пользователей и продажами средняя и равна', user_correl)
    if 0.7 < user_correl <= 1:
        print('Корреляция между оценкой пользователей и продажами высокая и равна', user_correl)
        
    if 0 < critic_correl <= 0.3:
        print('Корреляция между оценкой критиков и продажами слабая и равна', critic_correl)
    if 0.3 < critic_correl <= 0.7:
        print('Корреляция между оценкой критиков и продажами средняя и равна', critic_correl)
    if 0.7 < critic_correl <= 1:
        print('Корреляция между оценкой критиков и продажами высокая и равна', critic_correl)
    
```

## ШАГ 5: СОСТАВИМ ПОРТРЕТ ПОЛЬЗОВАТЕЛЯ В КАЖДОМ РЕГИОНЕ

```python
#Напишем функцию для создания топ-5 продаж по платформам, жанрам и рейтинку
def region_pivot_table(column, name):
    #Определим продажи во всем регионе
    total_column_sales = check_data[column].sum()
    #Делаем топ-5 по платформам
    platform_pivot_table = check_data.pivot_table(index='platform', values=column, 
                                                aggfunc='sum').sort_values(by=column, ascending=False)
    #Посчитаем долю продаж по платформе
    platform_pivot_table['sales_share'] = platform_pivot_table[column] / total_column_sales
    
    #делаем топ-5 по жанрам
    genre_pivot_table = check_data.pivot_table(index='genre', values=column, 
                                                aggfunc='sum').sort_values(by=column, ascending=False)
    #Считаем долю продаж по жанрам
    genre_pivot_table['sales_share'] = genre_pivot_table[column] / total_column_sales
    
    #делаем топ-5 по рейтингу
    rating_pivot_table = check_data.pivot_table(index='rating', values=column, 
                                                aggfunc='sum').sort_values(by=column, ascending=False)
    #Считаем долю продаж по рейтингу
    rating_pivot_table['sales_share'] = rating_pivot_table[column] / total_column_sales
    
    print('Самые популярные платформы в', name)
    display(platform_pivot_table.head())
    data_plot(platform_pivot_table, column, 'sales_share')
    print()
    print('Самые популярные жанры в', name)
    display(genre_pivot_table.head())
    data_plot(genre_pivot_table, column, 'sales_share')
    print()
    print('Самые популярные рейтинги в ', name)
    display(rating_pivot_table.head())
    data_plot(rating_pivot_table, column, 'sales_share')
```

```python
def data_plot(data,column1, column2):
    fig, ax = plt.subplots(1, 2, figsize=(18, 4))
    data.head().plot(y=column1, ax=ax[0], kind='bar', title='Распределение продаж по региону')
    data.head().plot(y=column2, ax=ax[1], kind='pie', title='Распределение продаж по региону в процентах')
    #fig.suptitle('Продажи и доли продаж')
    plt.show()
```

### Гипотеза 1 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

$H_o$ - нулевая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

$H_1$ - альтернативная гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC разные

Воспользуемся гипотезой о равенстве средних двух генеральных совокупностей

```python
#Задаем критический уровень статистической значимости
alpha = 0.05

xbox_one_hyp = check_data[(check_data['platform'] == 'XOne') & (check_data['year_of_release'] > 2009) & (check_data['user_score'] > 0)]['user_score']
pc_hyp = check_data[(check_data['platform'] == 'PC') & (check_data['year_of_release'] > 2009) & (check_data['user_score'] > 0)]['user_score']


varr_xone = np.var(xbox_one_hyp)
equal_var_xone = np.var(xbox_one_hyp, ddof=1)
print(varr_xone)
print(equal_var_xone)
variance = st.levene(xbox_one_hyp, pc_hyp)

print(variance)

varr_pc = np.var(pc_hyp)
equal_var_pc = np.var(pc_hyp, ddof=1)
print(varr_pc)
print(equal_var_pc)

results = st.ttest_ind(xbox_one_hyp, pc_hyp)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Не отвергаем нулевую гипотезу, так как так как уровень статистической значимости, который нужен для того, чтобы нулевая гипотеза выполнялась для полученного наблюдения высокая

$H_1$ - альтернативная гипотеза: Средние пользовательские рейтинги жанров Action и Sports разные

Воспользуемся гипотезой о равенстве средних двух генеральных совокупностей

```python
alpha = 0.05

action_hyp = check_data[(check_data['genre'] == 'Action') & (check_data['user_score'] > 1) & (check_data['year_of_release'] > 2009)]['user_score']
sports_hyp = check_data[(check_data['genre'] == 'Sports') & (check_data['user_score'] > 1) & (check_data['year_of_release'] > 2009)]['user_score']

results = st.ttest_ind(action_hyp, sports_hyp)



variance2 = st.levene(action_hyp, sports_hyp)
print(variance2)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Отвергаем нулевую гипотезу, так как уровень статистической значимости, который нужен для того, чтобы нулевая гипотеза выполнялась для полученного наблюдения почти 0
